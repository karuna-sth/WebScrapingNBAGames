```python
import pandas as pd
```

```python
df = pd.read_csv("scraped_data_nba_games.csv", index_col=0)
```

```python
df.shape
```

```python
df
```

```python
df = df.sort_values('date')
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]
```

```python
df
```

```python
df.columns.values.tolist()
```

```python
def add_target(team):
    team['target'] = team["won"].shift(-1)
    return team

df = df.groupby("teams", group_keys=False).apply(add_target)
```

```python
df[df["teams"] == 'LAL']
```

```python
df["target"][pd.isnull(df["target"])] = 2
```

```python
df["target"] = df["target"].astype(int, errors ="ignore")
df["won"] = df["won"].astype(int, errors ="ignore")
```

```python
df
```

```python
df["won"].value_counts()
```

```python
nulls = df.isnull().sum()
```

```python
nulls = nulls[nulls>0]
```

```python
nulls
```

```python
valid_columns = df.columns[~df.columns.isin(nulls.index)]
```

```python
df = df[valid_columns].copy()
```

```python
df
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

rr = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr, n_features_to_select=30, direction="forward", cv=split)
```

```python
removed_cols = ["season", "date", "won", "target", "teams", "teams_opp"]
```

```python

selected_cols = df.columns[~df.columns.isin(removed_cols)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[selected_cols] = scaler.fit_transform(df[selected_cols])
```

```python
df
```

```python
sfs.fit(df[selected_cols], df["target"])
```
